# Panorama Maker [Image Stitching]


```python
import cv2
import os
import numpy as np
from random import randint
import matplotlib.pyplot as plt
```

```python
# Define Function to Take _Shots ():
def Take_Shots():
    cap =cv2.VideoCapture(0)
    os.mkdir('panorama_project2')
    i = 0
    while(1):
        ret,frame = cap.read()  
        cv2.imshow('Live',frame)
        if cv2.waitKey(10) & 0xFF == ord('g'):  #" Press g to save the image then q to close"
            cv2.imwrite('panorama_project2/cLiveFrame'+str(i)+'.jpg',frame)
            i +=1
            print(i)
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break                                                                          
    cap.release()                             
    cv2.destroyAllWindows()
    # with random names and save them in certain folder
```

```python
Take_Shots()
```

```python
# Read Images from that folder
images_list=[]
for img in  os.listdir("panorama_project2"):
    image=cv2.imread("panorama_project2/"+img)
    plt.imshow(image)
    plt.show()
    images_list.append(image)
```

```python
stitcher = cv2.Stitcher.create()
ret,pano = stitcher.stitch(images_list)

if ret==cv2.STITCHER_OK:
    cv2.imshow('Panorama',pano)
    cv2.waitKey()
    cv2.destroyAllWindows()
else:
    print("Error during Stitching")
```
